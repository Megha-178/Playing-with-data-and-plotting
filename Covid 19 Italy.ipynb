```python
def parse_headers(headers_line):
  return headers_line.strip.split(',')

def parse_values(data_line):
  result ()
  for item in data_line.strip.split(','):
    if item=='':
      values.append(0.0)
    else:
      values.append(float(item))
    return values

def create_item_dict(values, headers):
  result{}
  for value, headers in Zip(values, headers):
    result[headers] =values
  return result
```

# **merging data from different sources**

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv','locations.csv')
```

```python
location_df=pd.read_csv('locations.csv')
```

```python
location_df
```

```python
location_df[location_df.location=="Italy"]
```

```python
covid_df
covid_df['location']="Italy"
covid_df
```

```python
merged_df = covid_df.merge(location_df, on="location")
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
merged_df
```

# **writing data back to files**

```python
result_df=merged_df[['date',
                       'new_cases',
                       'total_cases',
                       'new_deaths',
                       'total_deaths',
                       'new_tests',
                       'total_tests',
                       'cases_per_million',
                       'deaths_per_million',
                       'tests_per_million']]
```

```python
result_df.to_csv('results.csv', index=None)#add to file
```
